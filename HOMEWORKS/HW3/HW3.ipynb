We want to investigate:

- The so called pairs-plot is a very simple way of quickly analysing realtionships between data. Generate the following figure

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv("IRIS.csv")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))  # one row, two columns

# First scatter plot
sns.scatterplot(
    data = df1,
    x ="sepal_length",
    y ="petal_length",
    hue = "species",
    ax = axes[0] # put at position 1, here that is index 0 since python indexing
)

# Second scatter plot
sns.scatterplot(
    data = df1,
    x = "sepal_width",
    y = "petal_width",
    hue = "species",
    ax = axes[1] # put at position 2, here that is index 0 since python indexing
)

plt.show()
```

What we can tell from this figure is that the species iris-versicolor and iris-virginicia have a much wider petal length and width compared to iris-setosa with iris-virginica having the largest petals. 

As for the sepal length we can see that it is not as straightforward since we see iris setosa has biggest sepal width on average while iris versicolor and iris virginica having much longer sepals with virginica being on top. 

A possible hypothesis we could draw from this is that iris virginica and veriscolor are more closely related compared to iris setosa. We can also say there might be a sligt relationship between sepal dimensions and petal dimensions. 

```python
# Create a custom grid of subplots (2 rows, 4 columns)
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Example scatterplots on specific axes
sns.boxplot(
    data=df1,
    x="species",
    y="sepal_width", 
    ax=axes[0, 0],
    hue = "species"
)
sns.boxplot(
    data=df1, 
    x="species", 
    y="sepal_length", 
    ax=axes[0, 1],
    hue = "species"
)


sns.boxplot(
    data=df1, 
    x="species", 
    y="petal_width", 
    ax=axes[1, 0],
    hue = "species"
)
sns.boxplot(
    data=df1, 
    x="species", 
    y="petal_length", 
    ax=axes[1, 1],
    hue = "species"
)

plt.show()
```

To answer how the sepal and petal dimensions are distributed we can see that the median is roughly the expected value (middle of each respective boxplot) indicating a normal distribution among all the species which is a reasonable conclusion seeing how growth of many things in nature follows a normal distribution. 

```python
sns.pairplot(df1, hue="species", markers=["o", "s", "D"])
```

Can see that the petal length and width among iris setosa follows a normal distribution with a much lower expected value compared to the other two. We can however see that sepal length and width are much more alike among the three species. However, even so, among iris setosa we see the sepal length being smaller while sepal width being bigger.

# Predicting strokes

```python
df3 = pd.read_csv("stroke-data.csv")

df3.head(), df3.info()
```

## Do urban or rural populations differ in stroke occurrence?
Plot how many have been affected by stroke in the two populations and see which of the two seems to have a more prevalent risk of having a stroke. 

```python
df3_residence_cleaned = df3["Residence_type"].dropna()
df3_grouped = df3_residence_cleaned.value_counts() # count values for each type

df3_urban_strokes = df3_grouped.get("Urban", 0) # fetch all urban types and replace by 0 if NA
df3_rural_strokes = df3_grouped.get("Rural", 0) # fetch all rural types and replace by 0 if NA

residence_types = ["Urban", "Rural"]
stroke_counts = [df3_urban_strokes, df3_rural_strokes]

plt.bar(residence_types, stroke_counts)
plt.xlabel("Residence Type")
plt.ylabel("Stroke Count")
plt.title("Stroke Count by Residence Type")
plt.show()
```

We can see a relatively safe to assume insignificant difference between the two types of residences and the stroke count. 

## How do the factors presented in the dataset differ from urban/rural populations?
Scatterplot with line-average BMI/gender etc... for both of the populations and see how it differs.

```python
# clean data
df3_health_cleaned = df3.dropna(subset=[
    "hypertension", 
    "heart_disease", 
    "avg_glucose_level", 
    "bmi", 
    "smoking_status", 
    "stroke", 
    "Residence_type"
])

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle("Health Distributions and Socioeconomic Factors relative Stroke Count", fontsize=16, y=1.02)

# BMI distribution with stroke
sns.kdeplot(
    data=df3_health_cleaned, 
    x="bmi", 
    hue="stroke", 
    multiple="fill", 
    ax=axes[0, 0], 
    palette="coolwarm"
)
axes[0, 0].set_title("BMI Distribution")
axes[0, 0].set_xlabel("BMI")
axes[0, 0].set_ylabel("Density")

# average glucose distribution with stroke
sns.kdeplot(
    data=df3_health_cleaned, 
    x="avg_glucose_level", 
    hue="stroke", 
    multiple="fill", 
    ax=axes[0, 1], 
    palette="coolwarm"
)
axes[0, 1].set_title("Avg Glucose Level Distribution")
axes[0, 1].set_xlabel("Average Glucose Level (mg/dL)")
axes[0, 1].set_ylabel("Density")

# function to add count labels on countplot bars
def add_count_labels(ax):
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', fontsize=11, color='black', xytext=(0, 5), 
                    textcoords='offset points')

# hypertension count with stroke
hypertension_plot = sns.countplot(
    data=df3_health_cleaned, 
    x="hypertension", 
    hue="stroke", 
    ax=axes[0, 2], 
    palette="coolwarm"
)
axes[0, 2].set_title("Hypertension and Stroke Count")
axes[0, 2].set_xlabel("Hypertension (hadn't or had)")
axes[0, 2].set_ylabel("Count")
add_count_labels(hypertension_plot)

# heart disease count with stroke
heart_disease_plot = sns.countplot(
    data=df3_health_cleaned, 
    x="heart_disease", 
    hue="stroke", 
    ax=axes[1, 0], 
    palette="coolwarm"
)
axes[1, 0].set_title("Heart Disease and Stroke Count")
axes[1, 0].set_xlabel("Heart Disease (hadn't or had)")
axes[1, 0].set_ylabel("Count")
add_count_labels(heart_disease_plot)

# smoking status distribution with stroke
smoking_status_plot = sns.countplot(
    data=df3_health_cleaned, 
    x="smoking_status", 
    hue="stroke", 
    ax=axes[1, 1], 
    palette="coolwarm"
)
axes[1, 1].set_title("Smoking Status and Stroke Count")
axes[1, 1].set_xlabel("Smoking Status")
axes[1, 1].set_ylabel("Count")
axes[1, 1].tick_params(axis='x', rotation=45)
add_count_labels(smoking_status_plot)

# stroke by residence type
residence_type_plot = sns.countplot(
    data=df3_health_cleaned, 
    x="Residence_type", 
    hue="stroke", 
    ax=axes[1, 2], 
    palette="coolwarm"
)
axes[1, 2].set_title("Stroke by Residence Type")
axes[1, 2].set_xlabel("Residence Type")
axes[1, 2].set_ylabel("Count")
add_count_labels(residence_type_plot)

plt.tight_layout()
plt.show()
```

BMI plot showing that a healthy BMI (often said to be around 20 but differs slightly from source to source) starts to show some signs of increased risk of stroke as can be seen by the increase in the orange distribution. After BMI 40 there is a sudden spike in stroke count after which the count seems to decrease drastically which is likely a consequence of the early deahts caused by such a high BMI. 

Average glucose level plot showing that a drastically lower/high average glucose level (lower/higher than 60-120 mg/dL) gives birth to a rise in stroke count as can be seen by the spikes in orange. 

Hypertension seemingly not implying that having hypertension (1) implies a higher risk of having stroke due to the proportion of L/R column being roughly the same for both having and not having hypertension.

Heart disease the same as hypertension. 

Smoking is roughly the same. 

Residence type seems to have a small effect on the amount of strokes with urban having a larger stroke count. However, the total amount of people in urban is also larger and could therefore cause this to happen. 



## Do the different factors have a more significant effect on the possibility of stroke in urban/rural populations?
Plot heatmap of the different residental types. 

```python
# Clean data and ensure numeric columns
df3_health_cleaned = df3.dropna(subset=[
    "hypertension", 
    "heart_disease", 
    "avg_glucose_level", 
    "bmi", 
    "smoking_status", 
    "stroke", 
    "Residence_type"
])

# Convert binary categorical columns to numeric using `.loc[]`
df3_health_cleaned.loc[:, "hypertension"] = df3_health_cleaned["hypertension"].astype(int)
df3_health_cleaned.loc[:, "heart_disease"] = df3_health_cleaned["heart_disease"].astype(int)
df3_health_cleaned.loc[:, "stroke"] = df3_health_cleaned["stroke"].astype(int)
df3_health_cleaned.loc[:, "smoking_status"] = df3_health_cleaned["smoking_status"].astype("category").cat.codes

# Separate Urban and Rural populations
urban_data = df3_health_cleaned[df3_health_cleaned["Residence_type"] == "Urban"]
rural_data = df3_health_cleaned[df3_health_cleaned["Residence_type"] == "Rural"]

# Calculate correlation matrices, dropping non-numeric columns
numeric_columns = ["hypertension", "heart_disease", "avg_glucose_level", "bmi", "stroke", "smoking_status"]

urban_corr = urban_data[numeric_columns].corr()
rural_corr = rural_data[numeric_columns].corr()

# Create heatmaps for Urban and Rural correlations
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Urban correlation heatmap
sns.heatmap(
    urban_corr, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm", 
    ax=axes[0]
)
axes[0].set_title("Urban Population Correlation Matrix")

# Rural correlation heatmap
sns.heatmap(
    rural_corr, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm", 
    ax=axes[1]
)
axes[1].set_title("Rural Population Correlation Matrix")

# Adjust layout
plt.tight_layout()
plt.show()
```

Showcasing some differences in the correlationcoefficient in the stroke column between the two residence types. The differences are so small however that this barely actually means the factors have a smaller or larger correlation to having a stroke between the two residence types. 

# Cleaning data
- Deal with missing values in a suitable way (fill and/or remove missing values).
- Convert all the **relevant** cells to numbers (not strings).

```python
cell_phones_data = pd.read_csv("cell_phones_total.csv", delimiter = ",")
```

```python
cell_phones_data.head()
```

We see that many years have missing data. Inconsistent formats with suffixes. Inconsistent dtypes, all should be floats. 

```python
# convert to values and not just shortened indexes
def convert_shorthand(value):
    if isinstance(value, str):
        if 'k' in value:
            return float(value.replace('k', '')) * 1e3
        elif 'M' in value:
            return float(value.replace('M', '')) * 1e6
        elif 'B' in value:
            return float(value.replace('B', '')) * 1e9
    return value

# apply the conversion function to all columns except 'iso-3' since not a numerical
for column in cell_phones_data.columns[1:]:
    cell_phones_data[column] = cell_phones_data[column].apply(convert_shorthand)

# handle missing values: use forward fill and then backward fill to handle missing data | thank you chatgpt
cell_phones_data.ffill(axis=0, inplace=True)  # Forward fill
cell_phones_data.bfill(axis=0, inplace=True)  # Backward fill

# convert all columns (except 'iso-3') to numeric data types
for column in cell_phones_data.columns[1:]:
    cell_phones_data[column] = pd.to_numeric(cell_phones_data[column], errors='coerce')

# display the cleaned data starting from year 2015 onwards sorted in ascending order in the column 2015
cleaned_data_2015_onwards = cell_phones_data[['iso-3', '2015', '2016', '2017', '2018', '2019']]
cleaned_data_2015_onwards_sorted = cleaned_data_2015_onwards.sort_values(by="2015", ascending= False)

# Save the cleaned data to a new CSV file
output_file_path = "cleaned_cell_phone_data.csv"
cleaned_data_2015_onwards_sorted.to_csv(output_file_path, index=False)

output_file_path = "cleaned_cell_phone_data_all_years.csv"
cell_phones_data.to_csv(output_file_path, index=False)

print(f"Cleaned data has been saved to {output_file_path}")
```
