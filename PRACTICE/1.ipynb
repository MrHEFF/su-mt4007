```python
strokes = pd.read_csv(r'C:\Users\sebas\OneDrive\UNIVERSITY\HT 24\MT4007 - statistical data processing\PRACTICE\stroke-data.csv')
```

```python
strokes = strokes[['gender', 'age', 'smoking_status', 'heart_disease', 'bmi', 'stroke']]

strokes["bmi"].fillna(strokes["bmi"].mean(), inplace = True)

strokes["gender"] = strokes["gender"].map({
    'Male': 0, 
    'Female': 1, 
    'Other': 2
})

strokes['smoking_status'] = strokes['smoking_status'].map({
    'never smoked': 0,
    'formerly smoked': 1,
    'smokes': 2,
    'Unknown': 3
})


corr_matrix = strokes.corr()
corr_matrix

import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
